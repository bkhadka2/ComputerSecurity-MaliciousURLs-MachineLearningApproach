```python
import pandas as pd
import numpy as np
print('import complete')
```

```python
df = pd.read_csv('Cleaned_Data.csv')
```

```python
df.head()
```

```python
del df['Unnamed: 0']
```

```python
df.head()
```

```python
selectedFeatures = ['pathLength', 'subDirLen', 'charcompvowels', 'ldl_path', 'ldl_url', 'URL_sensitiveWord', 'isPortEighty', 'charcompace', 'dld_domain', 'NumberRate_Domain', 'URL_Type_obf_Type']
```

```python
for mem in df.columns:
    if mem not in selectedFeatures:
        del df[mem]
```

```python
df.head()
```

```python
charcompvowelsList = df['charcompvowels']
charcompaceList = df['charcompace']
ldl_urlList = df['ldl_url']
ldl_pathList = df['ldl_path']
dld_domainList = df['dld_domain']
pathLengthList = df['pathLength']
subDirLenList = df['subDirLen']
isPortEightyList = df['isPortEighty']
URL_sensitiveWordList = df['URL_sensitiveWord']
NumberRate_DomainList = df['NumberRate_Domain']
```

```python
combinedList = list(zip(charcompvowelsList, charcompaceList, ldl_urlList, ldl_pathList, dld_domainList, pathLengthList, subDirLenList, isPortEightyList, URL_sensitiveWordList, NumberRate_DomainList))
```

```python
df['URL_Type_obf_Type'] = df['URL_Type_obf_Type'].map(lambda val : 0 if val == 'benign' else 1)
```

```python
df.head()
```

```python
target_value = list()
```

```python
for mem in df['URL_Type_obf_Type']:
    target_value.append(mem)
```

```python
from sklearn.model_selection import train_test_split
```

```python
random_state = 1
X = combinedList
y = target_value
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=random_state, stratify=y)
```

```python
print("Number of training samples: {}".format(len(X_train)))
print("Number of testing samples: {}".format(len(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
randomForest = RandomForestClassifier(max_depth=2, random_state=random_state)
decisionTree = DecisionTreeClassifier(random_state = random_state)
```

```python
randomForest.fit(X_train, y_train)
print("Random Forest fitted")
decisionTree.fit(X_train, y_train)
print("Decision Tree fitted")
```

```python
randomForestResult = randomForest.score(X_test, y_test)
decisionTreeResult = decisionTree.score(X_test, y_test)

print('Random Forest: {:.2f}'.format(randomForestResult))
print('Decision Tree: {:.2f}'.format(decisionTreeResult))
```
